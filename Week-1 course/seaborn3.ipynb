```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.set_style('darkgrid')

sns.pairplot(tips);
```

```python
tips.dtypes
```

```python
sns.distplot(tips['size'],
            kde=False,
            hist_kws={'alpha':1});
```

```python
sns.pairplot(tips,
            #  hue='weekend',
            y_vars=['tip'],
            x_vars=['total_bill','size'],
             kind='reg',
             height=4);
```

```python
from ipywidgets import interactive

def size_widget(height=2.5, aspect=1):
    sns.pairplot(tips,height=height, aspect=aspect)
```

```python
interactive(size_widget,height=(1,3.5,0.5), aspect=(0.5,2, 0.25))
```

```python
sns.pairplot(tips, hue='day');
```

```python
sns.pairplot(tips, 
             kind='reg',
             plot_kws={
                 'ci':None,
                 'color': 'xkcd:salmon',
                 'scatter_kws':{'color':'gray'}
                     });
```

```python
penguins = sns.load_dataset('penguins')

penguins.head()
```

```python
penguins.shape
```

```python
sns.set_style('darkgrid')

sns.barplot(x='species',
            y='body_mass_g',
            data=penguins);
```

```python
penguins.groupby('species').body_mass_g.mean()
```

```python
sns.barplot(y='species',
            x='body_mass_g',
            data=penguins);
```

```python
sns.barplot(x='species',
            y='body_mass_g',
            data=penguins,
            ci=None);
```

```python
sns.barplot(y='species',
            x='body_mass_g',
            data=penguins,
            estimator=np.std,
            hue='sex');
```

```python
penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(x='species',
            y='body_mass_g',
            data=penguins,
            estimator=np.max,
            hue='sex');
```

```python
sns.barplot(x='species',
            y='body_mass_g',
            data=penguins,
            hue='sex',
            order=['Gentoo','Chinstrap', 'Adelie'],
            hue_order=['Female', 'Male'],
            palette='hot');
```
