```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.kdeplot(cars.horsepower,shade=True);
```

```python
sns.kdeplot(cars.horsepower,cumulative=True);
```

```python
sns.kdeplot(cars.horsepower,cars.mpg,shade=True);
```

```python
sns.boxplot(y='origin',
            x='mpg',
            data=cars,
            order=['japan' , 'europe' , 'usa'],
            hue_order=True,
            width=0.5,
            whis=1.5);
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds = diamonds[
    diamonds.cut.isin(['Premium', 'Good']) &
    diamonds.color.isin(['D', 'F', 'J'])
].sample(n=100, random_state=22)
diamonds.shape
```

```python
sns.scatterplot(x=diamonds.carat, y=diamonds.price);
```

```python
sns.scatterplot( x = 'carat',y = 'price',
                data = diamonds,
                palette=[]
                );
```

```python
sns.scatterplot( x = 'carat',y = 'price',
                data = diamonds,
                hue='depth'
                );
```

```python
sns.scatterplot( x = 'carat',y = 'price',
                data = diamonds,
                s=100
                );
```

```python
sns.scatterplot( x = 'carat',y = 'price',
                data = diamonds,
                style='cut'
                );
```

```python
sns.scatterplot( x = 'carat',y = 'price',
                data = diamonds,
                size='cut',
                
                );
```
